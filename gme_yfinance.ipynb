```python
import yfinance as yf

# Extracting GME Stock Data
gme_ticker = yf.Ticker("GME")
gme_data = gme_ticker.history(period="max")

# Resetting the index
gme_data_reset = gme_data.reset_index()

# Convert the DataFrame to JSON format; orient='records' makes the JSON file more readable
# by writing each row as a JSON object
gme_json = gme_data.to_json(orient='records')

# Define the filename
filename = 'gme.json'

# Write the JSON data to a file
with open(filename, 'w') as file:
    file.write(gme_json)

# Display the first five rows of the dataframe
gme_data_reset_head = gme_data_reset.head()
gme_data_reset_head
```
