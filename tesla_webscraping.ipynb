```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = 'https://stockanalysis.com/stocks/tsla/revenue/'
data  = requests.get(url).text

soup = BeautifulSoup(data, 'html5lib')

# Find the table containing the data.
# The class name used here is an example and needs to be replaced with the actual class name.
table = soup.find('table')

# Extract table rows.
rows = table.find_all('tr')

# Process each row and extract data.
data = []
for row in rows:
    cols = row.find_all('td')
    cols = [ele.text.strip() for ele in cols]
    data.append([ele for ele in cols if ele])

# Create a DataFrame and display the data.
df = pd.DataFrame(data)
df.columns = ["Year End", "Revenue", "Change", "Growth"]

# Remove the first row if it's not needed
df = df.drop(0)

# Reset the index if necessary
df = df.reset_index(drop=True)

df_head = df.tail()

# Apply the styling
styled_df = df_head.style.set_properties(**{
    'background-color': 'white', 
    'color': 'black', 
    'border-color': 'gray'
})

# You can also set the properties for specific columns, for example:
styled_df = styled_df.set_table_styles({
    'Date': [{'selector': '',
             'props': [('color', 'red')]}],
    # Add more columns as needed
}, overwrite=False)

# Display the styled DataFrame
styled_df
```
